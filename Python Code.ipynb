```python
import pandas as pd
import numpy as np

import warnings
warnings.filterwarnings('ignore')
```

```python
#reading the input file
df=pd.read_excel('input.xlsx')
df.head()
```

```python
#defining a function to get the text from the articles
from newspaper import Article
import nltk

def get_text(url):
    try:
        article = Article(url, language="en")
        article.download()
        article.parse()
        article.nlp()
        return article.text
    except Exception as e:
        print(f"Error fetching text for URL: {url}")
        return ""
```

```python
for i in range(0,len(df)):
    df['URL_ID'][i] = get_text(df['URL'][i])
```

```python
# I want a combined dataframe with url_id too
df2=pd.read_excel('input.xlsx')

data = pd.merge(df2, df, on='URL', how='inner')
```

```python
data
```

```python
# Now we will do some text preprocessing and cleaning
import re
from nltk.corpus import stopwords
def transform(text):

    review = re.sub('[^a-zA-Z0-9]', ' ',text)
    review = review.lower()
    review = review.split()
    
    review = [word for word in review if not word in stopwords.words('english')]
    review = ' '.join(review)
    return review


data['Transform_Text'] = data['Text'].apply(transform)
```

## Data Analysis

```python
# word count in each text row.

data['WORD COUNT'] = data['Transform_Text'].apply(lambda x: len(str(x).split())) 
```

```python
data['AVG NUMBER OF WORDS PER SENTENCE'] = np.nan

for i in range(0,len(data)):
    
    data['AVG NUMBER OF WORDS PER SENTENCE'][i] = data['WORD COUNT'][i]/len(nltk.sent_tokenize(data['Text'][i]))
```

```python
def char_count(x):
    s = x.split()
    x = ''.join(s)
    return len(x)
data['chara_count'] = data['Transform_Text'].apply(lambda x: char_count(x))
```

```python
data['AVG WORD LENGTH'] = np.nan

for i in range(0,len(data)):
    
    data['AVG WORD LENGTH'][i] = data['chara_count'][i]/data['WORD COUNT'][i]
```

```python
data.head(10)
```

```python
def syllable_count(x):
    v = []
    d = {}
    for i in x:
        if i in "aeiou":
            v.append(i)
            d[i] = d.get(i,0)+1
            
    k = []
    for i in d:
        k.append(d[i])

    return np.sum(k)

data['syllable count'] = data['Transform_Text'].apply(lambda x: syllable_count(x))
```

```python
from collections import  Counter
def complex_word_count(x):
    
    syllable = 'aeiou'
    
    t = x.split()
    
    v = []
    
    for i in t:
        words = i.split()
        c=Counter()
        
        for word in words:
            c.update(set(word))

        n = 0
        for a in c.most_common():
            if a[0] in syllable:
                if a[1] >= 2:
                    n += 1
                
        m = 0
        p = []
        for a in c.most_common():
            if a[0] in syllable:
                p.append(a[0])
        if len(p) >= 2:
            m += 1
        
        if n >= 1 or m >= 1:
            v.append(i)
            
    return len(v) 

data['COMPLEX WORD COUNT'] = np.nan

data['COMPLEX WORD COUNT'] = data['Transform_Text'].apply(lambda x: complex_word_count(x))
```

```python
data['SENTENCE LENGTH'] = np.nan
data['AVG SENTENCE LENGTH'] = np.nan
data['PERCENTAGE OF COMPLEX WORDS'] = np.nan
data['FOG INDEX'] = np.nan


for i in range(0,len(data)):
    
    data['SENTENCE LENGTH'][i]  =   len(nltk.sent_tokenize(data['Text'][i]))
    data['AVG SENTENCE LENGTH'][i] = data['WORD COUNT'][i]/data['SENTENCE LENGTH'][i]
    data['PERCENTAGE OF COMPLEX WORDS'][i] = data['COMPLEX WORD COUNT'][i]/data['WORD COUNT'][i] 
    data['FOG INDEX'][i] = 0.4 * (data['AVG SENTENCE LENGTH'][i] + data['PERCENTAGE OF COMPLEX WORDS'][i])
```

```python
# For Syllable per Word

data['SYLLABLE PER WORD'] = np.nan

for i in range(0, len(data)):
    
    data['SYLLABLE PER WORD'][i] = data['syllable count'][i]/data['WORD COUNT'][i]
```

```python
# Now to calculate Positive Scores
with open('positive-words.txt', 'r') as file:
    positive_words = [line.strip() for line in file]

def calculate_positive_score(sentence):
    words = sentence.split()
    positive_score = sum(1 for word in words if word.lower() in positive_words)
    return positive_score
data['POSITIVE SCORE'] = data['Transform_Text'].apply(calculate_positive_score)
```

```python
data.head()
```

```python
# Same for Negative Scores
with open('negative-words.txt', 'r') as file:
    negative_words = [line.strip() for line in file]

def calculate_negative_score(sentence):
    words = sentence.split()
    negative_score = sum(1 for word in words if word.lower() in negative_words)
    return negative_score
data['NEGATIVE SCORE'] = data['Transform_Text'].apply(calculate_negative_score)
```

```python
data.head()
```

```python
# For Polarity Score
data['POLARITY SCORE'] = np.nan
for i in range(len(data)):
    data['POLARITY SCORE'][i]=(data['POSITIVE SCORE'][i]-data['NEGATIVE SCORE'][i])/(data['POSITIVE SCORE'][i]+data['NEGATIVE SCORE'][i]+0.000001)
```

```python
data.head()
```

```python
# Now for Sunjectivity Score
from textblob import TextBlob
data['SUBJECTIVITY SCORE'] = np.nan

for i in range(len(data)):
    data['SUBJECTIVITY SCORE'][i] = TextBlob(data['Transform_Text'][i]).sentiment[1]
```

```python
data.head(10)
```

```python
# Now for the Personal Pronouns
pattern = r'\b(i|we|my|mine|ours|us)(?![A-Za-z])\b' 
# The negative lookahead (?![A-Za-z]) ensures that "US" is not included because 
# it checks that there are no alphabetic characters immediately following the pronoun.

def count_personal_pronouns(sentence):
    pronoun_matches = re.findall(pattern, sentence)
    pronoun_count = len(pronoun_matches)
    return pronoun_count

data['PERSONAL PRONOUNS'] = data['Transform_Text'].apply(count_personal_pronouns)
```
